```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import seaborn as sns
```

```python
# Uploading the data set to google colab
uploaded = files.upload()
```

```python
# Reading the csv and populating a data frame
file_csv = list(uploaded.keys())[0]
df = pd.read_csv(file_csv)
```

```python
# Describing the data
print(df.describe())
```

```python
# Showing the data
# df.head(10)
df[:10]
```

```python
# Data visulaization with scatter plot

# data
x = df["MonthlyIncome"]
y = df["YearsAtCompany"]


# Create scatter plot
plt.figure(figsize=(10,6))
plt.scatter(x, y, color="blue", edgecolor='k')

# Labels and title
plt.xlabel('MonthlyIncome')
plt.ylabel('YearsAtCompany')
plt.title('Scatter Plot : MonthlyIncome vs YearsAtCompany')


# Show plot
plt.show()
```

**Analysis of above scatter plot**: Employees with longer tenure generally earn higher salaries, but significant variability suggests other influencing factors like job role and performance.

```python
# Count attrition occurrences by Performance Rating using bar graph
attrition_counts = df[df["Attrition"] == "Yes"]["PerformanceRating"].value_counts().sort_index()

# Create bar graph
plt.figure(figsize=(8,5))
plt.bar(attrition_counts.index, attrition_counts.values, color="skyblue", edgecolor="black", width = 0.6)

# Labels and title
plt.xlabel("Performance Rating")
plt.ylabel("Attrition Count")
plt.title("Attrition Count by Performance Rating")
plt.xticks(attrition_counts.index)

# Show plot
plt.show()
```

**Analysis for the above graph**: Employees with low performance rate tend to loose their jobs.

```python
# Bar graph using sea born for Attrition by Job Role
plt.figure(figsize=(12,6))

# Create a count plot for Job Role with Attrition as hue
sns.countplot(x=df['JobRole'], hue=df['Attrition'])

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Labels and title
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.title('Attrition by Job Role')

# Display the graph
plt.show()
```

**Analysis for the above bar graph**: Some job roles experience higher attrition than others, particularly entry-level and technical positions.

```python
# Data visualisation using histogram
plt.figure(figsize=(10,5))

# Histogram for employees who left
sns.histplot(df[df['Attrition'] == 'Yes']['YearsAtCompany'], bins=10, kde=True, color='red')

# Labels and title
plt.xlabel('Years at Company')
plt.ylabel('Count of Employees')
plt.title('Histogram: Distribution of Years at Company for Attrition Cases')  # Plot title

# Show plot
plt.show()
```

**Analysis of the above plot:** The histogram reveals that employee attrition peaks around 3 years of tenure, suggesting a critical period for turnover. This indicates that employees may face challenges or dissatisfaction at this stage, leading to higher exits.

```python
# Data visualisation with a box plot
plt.figure(figsize=(18, 6))

# Box plot of MonthlyIncome by JobRole
sns.boxplot(x=df['JobRole'], y=df['MonthlyIncome'])

# Add labels and title
plt.xlabel('Job Role')
plt.ylabel('Monthly Income')
plt.title('Boxplot: Job Role vs. Monthly Income')

plt.show()
```

**Analysis of the above plot :** Higher-level jobs have greater median salaries and wider income ranges, while lower-level jobs have lower earnings with less variation.

```python
# Pie Chart: Attrition Rate Distribution

# Count the occurrences of each attrition status
attrition_counts = df['Attrition'].value_counts()

# Create a figure for the pie chart with a specified size
plt.figure(figsize=(6, 6))

# Create the pie chart with labels and percentage display
plt.pie(attrition_counts,
         labels=attrition_counts.index,
         autopct='%1.1f%%',
         colors=['lightcoral', 'lightskyblue'])

# Set the title of the pie chart
plt.title('Attrition Rate Distribution')

# Display the pie chart
plt.show()
```

Analysis: A small but significant proportion of employees leave each year, impacting company stability.

```python
# Hypothesis Testing

# T-test: Monthly Income and Attrition

from scipy.stats import ttest_ind

attrition_yes = df[df['Attrition'] == 'Yes']['MonthlyIncome']
attrition_no = df[df['Attrition'] == 'No']['MonthlyIncome']

t_stat, p_val = ttest_ind(attrition_yes, attrition_no, equal_var=False)
print(f'T-test results: t-statistic={t_stat:.2f}, p-value={p_val:.4f}')
```

Null Hypothesis (H0): There is no difference in monthly income between employees who left and those who stayed.

Alternative Hypothesis (H1): Employees who left had significantly lower incomes.

Result: A significant difference was found, indicating that lower-paid employees are more likely to leave.

**Answers for the Questions**

**Key Factors Affecting Attrition**: Low salary, poor performance rate, and shorter tenure contribute to higher attrition rates.

**Key Factors Affecting Satisfaction**: Higher salaries, good performance rate, and job stability improve satisfaction.

**Timing of Employee Exits**: Most employees leave within the first 3 years, indicating a need for early retention strategies.
